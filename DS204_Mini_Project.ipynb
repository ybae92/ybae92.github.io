```python
import ydata_profiling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Crate Profiling Report

```python
from ydata_profiling import ProfileReport
profile = ProfileReport(
    df,
    correlations={
        "pearson":     {"calculate": True},
        "spearman":    {"calculate": True},
        "kendall":     {"calculate": True},
        "phi_k":       {"calculate": False},
        "cramers":     {"calculate": False},
    }
)
profile.to_file("mini-project-happiness.html")
```

### Create Blot plot to check whether outliers exist or not.

```python

plt.figure(figsize=(4,2))
plt.title('Log GDP per capita')
sns.boxplot(x='Log GDP per capita', data=df)

q1 = df['Log GDP per capita'].quantile(0.25)
q3 = df['Log GDP per capita'].quantile(0.75)
iqr = q3-q1
upper = q3 + iqr*1.5
lower = q1 - iqr*1.5

print(f'upper: {upper:.2f}, lower: {lower:.2f}')
count_upper = df[df['Log GDP per capita'] > upper]
print(f'Count of high outlier : {count_upper['Log GDP per capita'].count()}')
count_lower = df[df['Log GDP per capita'] < lower]
print(f'Count of low outlier : {count_lower['Log GDP per capita'].count()}')

plt.tight_layout()
```
