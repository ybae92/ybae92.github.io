### Clean the data
1. clean the missing data
2. handle the outlier data

```python
clean_df = df.copy()
clean_df
```

#### 1. clean the missing data

```python
##Clean the missing data
##1. Log GDP per capita - Mean Imputation
clean_df['Log GDP per capita']=clean_df['Log GDP per capita'].fillna(clean_df['Log GDP per capita'].mean())
##2. Health life expectation - Median Imputation
clean_df['Healthy life expectancy at birth']=clean_df['Healthy life expectancy at birth'].fillna(clean_df['Healthy life expectancy at birth'].median())
##3. Freedom to make life choices - Median Imputation
clean_df['Freedom to make life choices']=clean_df['Freedom to make life choices'].fillna(clean_df['Freedom to make life choices'].median())
#4. Delete Row
for col in [
    'Perceptions of corruption',
    'Generosity',
    'Social support',
    'Negative affect',
    'Positive affect'
]:
  clean_df = clean_df.dropna(subset=[col])

```

```python
##Check after cleaning missing data
clean_df.isna().sum()
```

#### 2. Handle outliers


### Handling Outliers

#### Apply Log Transformation

```python
import numpy as np

# Social support - log transformation
# Add a small constant to avoid log(0) or log(negative) if there are any, though usually not an issue with social support scores.
clean_df['Social support'] = np.log(clean_df['Social support'])

# Healthy life expectancy at birth - log transformation
clean_df['Healthy life expectancy at birth'] = np.log(clean_df['Healthy life expectancy at birth'])

# Perceptions of corruption - log transformation
clean_df['Perceptions of corruption'] = np.log(clean_df['Perceptions of corruption'])

# Negative affect - log transformation
clean_df['Negative affect'] = np.log(clean_df['Negative affect'])

print('Log transformations applied to Social support, Healthy life expectancy at birth, Generosity, Perceptions of corruption, and Negative affect.')
```

#### Apply Clamping (Capping and Flooring) for 'Freedom to make life choices'

```python
# Freedom to make life choices - clamp
q1 = clean_df['Freedom to make life choices'].quantile(0.25)
q3 = clean_df['Freedom to make life choices'].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + iqr * 1.5
lower_bound = q1 - iqr * 1.5

clean_df['Freedom to make life choices'] = np.where(
    clean_df['Freedom to make life choices'] > upper_bound, upper_bound,
    np.where(clean_df['Freedom to make life choices'] < lower_bound, lower_bound,
             clean_df['Freedom to make life choices']
    )
)

print(f"Clamping applied to 'Freedom to make life choices' with upper bound: {upper_bound:.2f} and lower bound: {lower_bound:.2f}")
```

The average `estimation_error` represents the *mean bias* of the estimation. To understand the *magnitude* or 'rate' of the error, we can calculate the **Mean Absolute Error (MAE)**. MAE measures the average magnitude of the errors, regardless of their direction, providing a straightforward measure of how large the errors are on average.

```python
print('Average (Mean Bias) of estimation_error across groups:')
display(ab_test_df.groupby('group')['estimation_error'].mean())

print('\nMean Absolute Error (MAE) of estimation_error across groups:')
mae_estimation_error = ab_test_df.groupby('group')['estimation_error'].apply(lambda x: x.abs().mean())
display(mae_estimation_error)
```
